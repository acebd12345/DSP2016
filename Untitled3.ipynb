```python
import numpy as np
import pylab as pl
import pandas as pd
import thinkplot as tp
%matplotlib inline
f=open('coindesk-bpi-USD-close.csv')
s=f.read()
sL=s.split('\n')
xL=[x.split(',')[-1] for x in sL]
yL=[float(x) for x in xL[1:-4]]
priceData=np.array(yL)
```

```python
len(priceData)
```

```python
import thinkdsp as td
def movingAverage(x, length):
    y= np.convolve(x, np.ones(length)/length)
    y= y[:len(x)]
    return y

ma100= movingAverage(priceData, 100)
ma500= movingAverage(priceData, 500)
ma1000= movingAverage(priceData, 1000)
pl.plot(priceData)
pl.plot(ma100)
pl.plot(ma500)
pl.plot(ma1000)
```

```python
df = pd.read_csv('coindesk-bpi-USD-close.csv', nrows=2112, parse_dates=[0])
ys = df.Close.values
ts = np.arange(len(ys))
```

```python
wave = td.Wave(ys, ts, framerate=1)
wave.plot()
tp.plot.config(xlabel='Time (days)')
```

```python
spectrum = wave.make_spectrum()
priceSpec=spectrum
spectrum.plot_power()
tp.config(xlabel='Frequency (1/days)',
                 xscale='log', yscale='log')
```

```python
signal = td.BrownianNoise()
wave = signal.make_wave(duration=0.5, framerate=2112)
wave.make_audio()
```
