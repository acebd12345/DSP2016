```python
import numpy as np
import pylab as pl
import pandas as pd
import thinkplot as tp
import thinkdsp as td
%matplotlib inline
```

```python
signal = td.Chirp(start=100, end=1000)
wave1 = signal.make_wave(duration=10)
wave1.make_audio()
```

```python

wave1.segment(start=0, duration=0.2).plot()
```

```python
signal = td.Chirp(start=100, end=1000)
wave = signal.make_wave(duration=1)
spectrum = wave.make_spectrum()
spectrum.plot(high=1100)
tp.config(xlabel='frequency (Hz)')
```

```python
def plot_spectrogram(wave, seg_length):
    spectrogram = wave.make_spectrogram(seg_length)
    print('Time resolution (s)', spectrogram.time_res)
    print('Frequency resolution (Hz)', spectrogram.freq_res)
    spectrogram.plot(high=1100)
    tp.config(xlabel='Time(s)', ylabel='Frequency (Hz)')
```

```python
signal = td.Chirp(start=100, end=1000)
wave = signal.make_wave(duration=10, framerate=11025)
plot_spectrogram(wave, 1024)
```
