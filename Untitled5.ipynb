```python
import numpy as np
import pylab as pl
import pandas as pd
import thinkplot as tp
import thinkdsp as td
%matplotlib inline
```

```python
sin_sig = td.SinSignal(freq=100, amp=1, offset=0)
sin_sig.plot()
```

```python
signal = td.TriangleSignal(100, 1)
duration = signal.period*3
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
```

```python
wave = signal.make_wave(duration=0.5, framerate=10000)
spectrum = wave.make_spectrum()
spectrum.plot()
```

```python
signal = td.SquareSignal(100)
duration = signal.period*3
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
```

```python
wave = signal.make_wave(duration=0.5, framerate=10000)
spectrum = wave.make_spectrum()
spectrum.plot()
```
